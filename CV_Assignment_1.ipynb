```python
# convarting the image to gray
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("\nGray Image\n")   
cv2_imshow(grayImg)
# 4 connectivity
np.seterr(over='ignore')
[x,y] = grayImg.shape
newImg = copy.deepcopy(grayImg)
threshold = 10
for i in range(x):
  for j in range(y):
    if(i+1<x and abs(newImg[i+1,j] - newImg[i,j])<threshold): 
      newImg[i+1,j] = newImg[i,j]
    if(i-1>=0 and abs(newImg[i-1,j] - newImg[i,j])<threshold):
      newImg[i-1,j] = newImg[i,j]
    if(j+1<y and abs(newImg[i,j+1] - newImg[i,j])<threshold):
      newImg[i,j+1] = newImg[i,j]
    if(j-1>=0 and abs(newImg[i,j-1] - newImg[i,j])<threshold): 
      newImg[i,j-1] = newImg[i,j]
print("\n4 Connectivity\n")   
cv2_imshow(newImg)
# 4 connectivity diagonal ((x+1,y+1),(x+1,y-1),(x-1,y+1),(x-1,y-1))
newImg = copy.deepcopy(grayImg)
for i in range(x):
  for j in range(y):
    if((i+1<x and j+1 <y) and abs(newImg[i+1,j+1] - newImg[i,j])<threshold): 
      newImg[i+1,j] = newImg[i,j]
    if((i+1<x and j-1>=0) and abs(newImg[i+1,j-1] - newImg[i,j])<threshold):
      newImg[i-1,j] = newImg[i,j]
    if((i-1>=0 and j+1<y) and abs(newImg[i-1,j+1] - newImg[i,j])<threshold):
      newImg[i,j+1] = newImg[i,j]
    if((j-1>=0 and i-1>=0) and abs(newImg[i-1,j-1] - newImg[i,j])<threshold): 
      newImg[i,j-1] = newImg[i,j]

print("\n4 Connectivity Diagonal\n")      
cv2_imshow(newImg)
# 8 connectivity
newImg = copy.deepcopy(grayImg)
for i in range(x):
  for j in range(y):
    if(i+1<x and abs(newImg[i+1,j] - newImg[i,j])<threshold): 
      newImg[i+1,j] = newImg[i,j]
    if(i-1>=0 and abs(newImg[i-1,j] - newImg[i,j])<threshold):
      newImg[i-1,j] = newImg[i,j]
    if(j+1<y and abs(newImg[i,j+1] - newImg[i,j])<threshold):
      newImg[i,j+1] = newImg[i,j]
    if(j-1>=0 and abs(newImg[i,j-1] - newImg[i,j])<threshold): 
      newImg[i,j-1] = newImg[i,j]
    if((i+1<x and j+1 <y) and abs(newImg[i+1,j+1] - newImg[i,j])<threshold): 
      newImg[i+1,j] = newImg[i,j]
    if((i+1<x and j-1>=0) and abs(newImg[i+1,j-1] - newImg[i,j])<threshold):
      newImg[i-1,j] = newImg[i,j]
    if((i-1>=0 and j+1<y) and abs(newImg[i-1,j+1] - newImg[i,j])<threshold):
      newImg[i,j+1] = newImg[i,j]
    if((j-1>=0 and i-1>=0) and abs(newImg[i-1,j-1] - newImg[i,j])<threshold): 
      newImg[i,j-1] = newImg[i,j]

print("\n8 Connectivity\n")
cv2_imshow(newImg)
```
